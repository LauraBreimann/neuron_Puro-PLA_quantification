```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# specify the datasets for your plot
df = pd.read_csv('/Users/laurabreimann/Desktop/Inga/database/total_database.csv') 
```

```python
df.head()
```

```python
sns.reset_defaults()

# Create figure instance
fig, ax = plt.subplots(figsize=(4.5, 5))


#removing top and right borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


# Add boxplot
sns.boxplot(x=df['shRNA'],y=df['smFISH_per_area'],
hue=df['shRNA'], palette=["m", "g"])


sns.stripplot(x=df['shRNA'],y=df['smFISH_per_area'], hue=df['shRNA'], dodge=True)

plt.ylabel('smFISH per area', fontsize=15) 
plt.title('mRNA per area in neurites')
#plt.ylim(0.8, 1)
#plt.xlim(-0.1,1.7)
plt.xticks(fontsize=10)
 

#  Show and save the figure
plt.tight_layout()
plt.savefig('smFISH_per_area.png', dpi=300)
plt.show()
```
