```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# specify the datasets for your plot
df = pd.read_csv('/Users/laurabreimann/Desktop/Inga/database/total_database.csv') 
```

```python
df.head()
```

```python
sns.reset_defaults()

# Create figure instance
fig, ax = plt.subplots(figsize=(4.5, 5))


#removing top and right borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


# Add boxplot
sns.boxplot(x=df['shRNA'],y=df['smFISH_per_area'],
hue=df['shRNA'], palette=["m", "g"])


sns.stripplot(x=df['shRNA'],y=df['smFISH_per_area'], hue=df['shRNA'], dodge=True)

plt.ylabel('actin-ß-Puro-PLA per area', fontsize=15) 
plt.title('translation events in neurites')
plt.ylim(0, 0.1)
#plt.xlim(-0.1,1.7)
plt.xticks(fontsize=10)
 

#  Show and save the figure
plt.tight_layout()
plt.savefig('smFISH_per_area.png', dpi=300)
plt.show()
```

```python
Sample1 = df.loc[df['shRNA'] == 'ctrl']
Sample2 = df.loc[df['shRNA'] == 'shLarp']
```

```python
t, p = stats.ttest_ind(Sample1['smFISH_per_area'],Sample2['smFISH_per_area'], equal_var = True)
print("t-statistic of "  + str(np.unique(Sample1['smFISH_per_area'])) + " vs "+ str(np.unique(Sample2['smFISH_per_area']))+ "= " + str(t))
print("p-value of "  + str(np.unique(Sample1['smFISH_per_area'])) + " vs "+ str(np.unique(Sample2['smFISH_per_area']))+ " = " + str(p))
```
